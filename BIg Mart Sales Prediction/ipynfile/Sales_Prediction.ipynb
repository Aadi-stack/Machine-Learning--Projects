```python
y=df_train.iloc[:,-1:]
y
```

```python
X=df_train.iloc[:,0:9]
X
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.2) 
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
from sklearn.compose import make_column_transformer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import make_pipeline
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
ohe=OneHotEncoder()
one=OrdinalEncoder()
```

```python
ohe.fit(X[['Item_Fat_Content','Item_Type','Outlet_Location_Type']])
```

```python
one.fit(X[['Outlet_Size','Outlet_Type']])
```

```python
ohe.categories_
```

```python
one.categories_
```

```python
column_trans= make_column_transformer((OneHotEncoder(categories=ohe.categories_),['Item_Fat_Content','Item_Type','Outlet_Location_Type']),(OrdinalEncoder(categories=one.categories_),['Outlet_Size','Outlet_Type']),remainder='passthrough')
```

```python
column_trans
```

```python
lr=LinearRegression()
model1 = XGBRegressor()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,Y_train)
```

```python
y_pred=pipe.predict(X_test)
```

```python
y_pred
```

```python
r2_score(Y_test,y_pred)
```

```python
pipe=make_pipeline(column_trans,model1)
```

```python
pipe.fit(X_train,Y_train)
```

```python
y_pred1=pipe.predict(X_test)
```

```python
y_pred1
```

```python
r2_score(Y_test,y_pred1)
```

```python
# For XGBRegressor
RSQscore1 = metrics.r2_score(Y_test,y_pred1)

# Let's check

# The closer the errors are to 0, the more accurate our model is.
print("R squared error for XGB Regressor:",RSQscore1) 
```

```python
scores=[]
for i in range(1000):
    X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.1,random_state=i)
    lr=LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,Y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(Y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.1,random_state=np.argmax(scores))
lr=LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,Y_train)
y_pred=pipe.predict(X_test)
r2_score(Y_test,y_pred)
```

```python
scores=[]
for i in range(1000):
    X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=0.1,random_state=i)
    model1 = XGBRegressor()
  
    pipe=make_pipeline(column_trans,model1)
    pipe.fit(X_train,Y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(Y_test,y_pred))
```
